```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot
```

```python
comparison=pd.read_excel("Comparison.xlsx")
```

```python
pyplot.figure(figsize=(10, 8))
figure=sns.heatmap(comparison,annot=True)
figure.set_yticklabels(['Random Forest','K Nearest Neighbour','Support Vector Machine','VGG16','ResNet50'],rotation=360)
pyplot.savefig('comparison.png', bbox_inches='tight')
pyplot.show()
```
